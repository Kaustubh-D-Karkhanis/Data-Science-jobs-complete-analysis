```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/ds_salaries.csv')
print(df)
```

```python
print(df['employment_type'].value_counts())
print(df['employment_type'].value_counts(normalize=True)*100)
```

```python
print(df['job_title'].value_counts(normalize=True)*100)
```

```python
df.loc[df['job_title']=='Data Scientist','salary_in_usd'].mean()
```

```python
avgsalaries = df.groupby('job_title')['salary_in_usd'].mean()
sorted_salaries=avgsalaries.sort_values(ascending=False)
top10=sorted_salaries.head(10)
```

```python
title_count=df['job_title'].value_counts()
title_count.head(10).plot.bar(color='yellow',xlabel='Job titles',ylabel='Number of jobs',title='Most number of job opportunities')
```

```python
top10.plot.bar(color='red')
plt.xlabel('Job Titles')
plt.ylabel('Average salary in USD')
plt.title('Top 10 job titles with highest average salary(USD)')
plt.show()
```

```python
employees_by_country=df['employee_residence'].value_counts()
employees_by_country.head(10).plot.pie(legend=True,title='Country wise distribution of data science employees')
```

```python
year_wise_experience_level=df.groupby(['work_year','experience_level']).size().unstack()
legends=['Entry level','Executive level/Director','Junior Mid level','Senior level']
ax=year_wise_experience_level.plot(kind='bar',stacked=False,title='Year wise experience levels of employees',xlabel='Number of employees',ylabel='Nummber of employees')
ax.legend(legends, title='Experience levels')
```
